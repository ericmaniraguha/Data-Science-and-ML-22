```python
import pandas as pd
```

```python
dataset_df1 =pd.DataFrame( { 'ID': [1, 2, 3, 5, 9],
                'Col-1': [1,2,3,4,5],
                'Col-2': [6,7,8,9,10],
                'Col-3': [11,12,13,14,15],
                'Col-4': ['apple','orange','banana','strawberry','rasberry'],  
}) 

dataset_df2 = pd.DataFrame( { 'ID': [1,1,3,5],
                'Col-A': [8,9,10,11],
                'Col-B': [12,13,15,17],
                'Col-4': ['apple','orange','banana','kiwi'],  
}) 
```

```python
dataset_df1
```

```python
dataset_df2
```

#### pd.merge()

* For combining data on common columns
* Most flexible, but also complex of the methods we will display 
* many to one  and many to many join are possible 
* side by side merge 

```python
inner = pd.merge(dataset_df1, dataset_df2)
inner
```

```python
# merge by specifying the column
pd.merge(dataset_df1, dataset_df2, on = 'ID')
```

```python
# Merge by specifying columns, it has sam results as we are merging columns that share names between dataframes
pd.merge(dataset_df1, dataset_df2, on = ['ID', 'Col-4'])
```

```python
# suffix and merging on columns that are unique tp each dataframe

pd.merge(dataset_df1, dataset_df2, suffixes= ['_l','_r'], left_on='Col-2', right_on='Col-A')
```

```python
# joining in indeces
pd.merge(dataset_df1, dataset_df2, suffixes= ['_l','_r'], left_index = True, right_index = True)
```

#### Type of Join

Inner Join  ===> pd.merge(df1, df2, on = 'col_name')

Left Join ===> pd.merge(df1, df2, on = 'col_name', how = 'left')

Right Join ===> pd.merge(df1, df2, on = 'col_name', how = 'right')

Outer Join ===> pd.merge(df1, df2, on = 'col_name', how = 'outer')
